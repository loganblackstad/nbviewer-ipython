```python
# plotting vertical plots in matplotlib

fig, axs = plt.subplots(3)
fig.suptitle('Vertically Stacked Trigonometric Subplots')

x = np.linspace(-np.pi*2, np.pi*2, 1001)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

axs[0].plot(x, y1)
axs[1].plot(x, y2)
axs[2].plot(x, y3)

axs[0].set(xlabel='Angle [rad]', ylabel='sin(x)')
axs[1].set(xlabel='Angle [rad]', ylabel='cos(x)')
axs[2].set(xlabel='Angle [rad]', ylabel='tan(x)', ylim=[-5,5])

for i in [0,1,2]: 
    axs[i].axhline(y=0, color='k', linewidth = .3)
    axs[i].axvline(x=0, color='k', linewidth = .3)

plt.tight_layout()

# adjust the plot spacing
plt.subplots_adjust(left=None, bottom=None, right=None, top=.9, wspace=None, hspace=None)

# left  = 0.125  # the left side of the subplots of the figure
# right = 0.9    # the right side of the subplots of the figure
# bottom = 0.1   # the bottom of the subplots of the figure
# top = 0.9      # the top of the subplots of the figure
# wspace = 0.2   # the amount of width reserved for blank space between subplots
# hspace = 0.2   # the amount of height reserved for white space between subplots
```

```python
# plotting horizontal plots in matplotlib 
# To obtain side-by-side subplots, pass parameters 1, 2 for one row and two columns.

fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
fig.suptitle('Horizontally Positioned Trigonometric Subplots')

x = np.linspace(-np.pi*2, np.pi*2, 1001)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)

ax1.set(xlabel='Angle [rad]', ylabel='sin(x)')
ax2.set(xlabel='Angle [rad]', ylabel='cos(x)')
ax3.set(xlabel='Angle [rad]', ylabel='tan(x)', ylim=[-5,5])

for i in [ax1,ax2,ax3]: 
    i.axhline(y=0, color='k', linewidth = .3)
    i.axvline(x=0, color='k', linewidth = .3)

plt.tight_layout()

# adjust the plot spacing
plt.subplots_adjust(left=None, bottom=None, right=None, top=.9, wspace=None, hspace=None)
```

```python
# plotting vertical plots in matplotlib using degrees 

fig, axs = plt.subplots(3)
fig.suptitle('Vertically Stacked Algebraic Subplots')

x = np.linspace(-180*2, 180*2, 1001)

y1 = np.sin(np.radians(x))
y2 = np.cos(np.radians(x))
y3 = np.tan(np.radians(x))

axs[0].plot(x, y1)
axs[1].plot(x, y2)
axs[2].plot(x, y3)

axs[0].set(xlabel='Angle [degrees]', ylabel='sin(x)')
axs[1].set(xlabel='Angle [degrees]', ylabel='cos(x)')
axs[2].set(xlabel='Angle [degrees]', ylabel='tan(x)', ylim=[-5,5])

for i in [0,1,2]: 
    axs[i].axhline(y=0, color='k', linewidth = .3)
    axs[i].axvline(x=0, color='k', linewidth = .3)

plt.tight_layout()

# adjust the plot spacing
plt.subplots_adjust(left=None, bottom=None, right=None, top=.9, wspace=None, hspace=None)

```
